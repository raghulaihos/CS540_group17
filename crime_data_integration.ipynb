```python
def create_master(name):

    path_city = path / name
    files_city = [file for file in path_city.iterdir()]
    data_city = []

    for file in files_city:
        if "lock" not in str(file):
            
            data_city.append(pd.read_csv(file))
            print(file)
    
    #print(data_city[0].columns)
    #print(data_city[1].columns == data_city[1].columns)
    
    
    data_master = None

    for d in data_city:
        if data_master is  None:
            data_master = d
        else:
            data_master = pd.concat([data_master,d])
    file_name = name + ".csv"
    data_master.to_csv(file_name)
    return data_master
```
