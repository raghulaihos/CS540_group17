```python
aqi = pd.read_csv('aqi_data.csv')
flights = pd.read_csv('flights_data.csv')
gas = pd.read_csv('gasoline_data.csv', thousands=',')
pb = pd.read_csv('public_transport_data.csv')
cities = pd.read_csv('uscities_data.csv')
```

```python
aqi = pd.read_csv('aqi_data.csv', parse_dates=['Date'])
aqi = aqi.drop(columns=["Unnamed: 0", 'State Code', 'County Code', 'Defining Site', 'Number of Sites Reporting'])
aqi['Year'] = aqi.Date.dt.year
aqi = aqi.drop(columns=["Date"])
aqi.to_csv('cleaned_aqi.csv')
```

```python
flights = flights.drop(columns=['Unnamed: 0', 'SEATS',
       'UNIQUE_CARRIER', 'ORIGIN_AIRPORT_ID',
       'ORIGIN_AIRPORT_SEQ_ID', 'ORIGIN_CITY_MARKET_ID',
       'ORIGIN_WAC', 'DEST_AIRPORT_ID', 'DEST_AIRPORT_SEQ_ID',
       'DEST_CITY_MARKET_ID',
       'DEST_STATE_FIPS',
       'DEST_WAC', 'AIRCRAFT_TYPE', 'Unnamed: 32', 'ORIGIN_STATE_FIPS'])
flights.to_csv('cleaned_flights.csv', index=False)
flights = pd.read_csv('cleaned_flights.csv')
flights.to_csv('cleaned_flights.csv', index=True)
flights["ORIGIN_CITY"] = flights["ORIGIN_CITY_NAME"].str.split(', ').str[0]
flights["DEST_CITY"] = flights["DEST_CITY_NAME"].str.split(', ').str[0]
flights = flights.drop(columns=['ORIGIN_CITY_NAME','DEST_CITY_NAME'])
flights = flights[['PASSENGERS', 'DISTANCE', 'AIR_TIME', 'UNIQUE_CARRIER_NAME',
       'ORIGIN','ORIGIN_CITY',
       'ORIGIN_STATE_ABR', 'ORIGIN_STATE_NM', 'ORIGIN_COUNTRY',
       'ORIGIN_COUNTRY_NAME', 'DEST', 'DEST_CITY', 'DEST_STATE_ABR', 'DEST_STATE_NM',
       'DEST_COUNTRY', 'DEST_COUNTRY_NAME', 'YEAR', 'QUARTER', 'MONTH']]
flights.to_csv('cleaned_flights.csv', index=True)
```

```python
pb = pb.drop(columns=['Legacy NTD ID', 'NTD ID', 'Organization Type',
       'Reporter Type', 'Primary UZA Population', 'VOMS'])
pb.to_csv('cleaned_pb.csv', index=False)
pb.columns
pb['bus'] = 
['bus'] = pb['Bus'] + pb['Articulated Bus'] + pb['Over-the-road Bus'] + pb['Double Decker Bus'] + pb['School Bus']
pb# = pb.drop(columns=['Bus', 'Articulated Bus', 'Over-the-road Bus',
       'Double Decker Bus', 'School Bus'])
pb['rail'] = pb['Heavy Rail Passenger Car'] + pb['Light Rail Vehicle'] + pb['Commuter Rail Passenger Coach'] + pb['Commuter Rail Self-Propelled Passenger Car']
pb = pb.drop(columns=['Heavy Rail Passenger Car',
       'Light Rail Vehicle', 'Commuter Rail Passenger Coach',
       'Commuter Rail Self-Propelled Passenger Car'])
pb = pb[['Name', 'City', 'State', 'bus', 'rail', 'Van', 'Cutaway', 'Automobile', 'Minivan',
       'Sports Utility Vehicle', 'Trolleybus', 'Locomotive',
       'Automated Guideway Vehicle', 'Vintage /Historic Trolley', 'Streetcar',
       'Aerial Tram', 'Monorail', 'Cable Car', 'Inclined Plane', 'Ferryboat',
       'Other', 'Total']]
pb = pb.drop(columns=["other"])
pb["other"] = pb['Other'] + pb['Sports Utility Vehicle'] + pb['Automated Guideway Vehicle'] + pb['Vintage /Historic Trolley'] + pb['Aerial Tram'] + pb['Inclined Plane']
pb = pb.drop(columns=['Other','Sports Utility Vehicle','Automated Guideway Vehicle','Vintage /Historic Trolley','Aerial Tram','Inclined Plane'])
pb = pb[['Name', 'City', 'State', 'bus', 'rail', 'Van', 'Cutaway', 'Automobile',
       'Minivan', 'Trolleybus', 'Locomotive', 'Streetcar', 'Monorail',
       'Cable Car', 'Ferryboat', 'other', 'Total']]
```

```python
cities = cities.drop(columns=['city', 'county_fips', 'county_fips_all', 'county_name_all', 'source', 'military', 'incorporated','ranking', 'id'])
cities.to_csv('cleaned_cities.csv', index=False)
```

```python
states = list(cities['state_name'].unique())
sflights = {}
saqi = {}
#for s in states:
#    sflights[s] = 0

for state in states:
    c = flights.loc[(flights['ORIGIN_STATE_NM'] == state) & (flights['YEAR'] > 2014)].size
    sflights[state] = c
    c = flights.loc[(flights['DEST_STATE_NM'] == state) & (flights['YEAR'] > 2014)].size
    sflights[state] += c
    
    saqi[state] = aqi.loc[aqi["State Name"].str.lower() == state.lower()]["AQI"].mean()
#sflights
```

```python
sgas = {}
for state in states:
    v = gas.loc[gas["State"].str.lower() == state.lower()]["Expenditures_mln_dlr"].mean()
    if not pd.isnull(v):
        sgas[state] = v
#import locale
#locale.atoi(gas.Expenditures_mln_dlr)
print("\n".join([str(k) + ": " + str(v) for k,v in sorted(sgas.items(), key=operator.itemgetter(1))[-10:]]))
```

```python
state_ids = list(cities['state_id'].unique())
spt = {}
for state in state_ids:
    #print(state, pb.loc[pb["State"].str.lower() == state.lower()]["Total"].sum())
    v = pb.loc[pb["State"].str.lower() == state.lower()]["Total"].sum()
    if not pd.isnull(v):
        spt[state] = v

print("\n".join([str(k) + ": " + str(v) for k,v in sorted(spt.items(), key=operator.itemgetter(1))[-10:]]))
```
